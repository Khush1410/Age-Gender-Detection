```python
img1=cv2.imread("image1.jpeg")
img=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV) # converts bgr to hsv (hue,saturation,value)
result=cv2.blur(img,(21,21))
plt.figure(figsize=(11,6))
plt.subplot(131),plt.imshow(cv2.cvtColor(img,cv2.COLOR_HSV2RGB)),plt.title("ORIGINAL")
plt.xticks([]),plt.yticks([]) # to remove x and y ticks from the axis
plt.subplot(132),plt.imshow(cv2.cvtColor(result,cv2.COLOR_HSV2RGB)),plt.title("BLURRED")
plt.xticks([]),plt.yticks([])
plt.imshow(img1)
```

```python
(height,width)=img1.shape[0:2]
```

```python
zoom=cv2.resize(img1,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
```

```python
rotate=cv2.warpAffine(img1,m,(height,width))
```

```python
plt.imshow(rotate)
```
